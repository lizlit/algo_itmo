```python
from timeit import default_timer as d_timer
import numpy as np
import sorting
```

```python
quickSort_time = np.zeros(2000)
def Quicksort(vector):
    return sorting.quick(vector)

def Const(vector):
    return 5.5

const_time = np.zeros(250)

for i in range(250):
    vector = np.random.rand(i+1)
    t0 = d_timer()
    result = Const(vector)
    const_time[i] = d_timer() - t0
```

```python
quickSort_time1 = np.zeros(2000)
for i in range(2000):
    vector = np.random.rand(i+1)
    t0 = d_timer()
    result = Quicksort(vector)
    quickSort_time1[i] = d_timer() - t0
```

```python
t_const = np.min(const_time)
theor_nlogn = np.zeros(2000)
def n_logn(x):
    return t_const * (x * np.log2(x))
```

```python
from matplotlib import pyplot as plt
```

```python
for i in range(2000):
    theor_nlogn[i] = n_logn(i)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(quickSort_time1)
plt.plot(2.6 * theor_nlogn)
plt.legend(['Empirical data', "Theor complexity Quick Sort "])
plt.savefig('quick.png')
```
